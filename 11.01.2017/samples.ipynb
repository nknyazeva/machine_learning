```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
samples = {}

def plot_sample(samples, name):
    plt.figure(figsize=(4, 4))
    plt.scatter(*samples[name])
    plt.title(name)
    plt.tick_params(bottom='off', top='off', left='off', right='off',
                    labelbottom='off', labelleft='off')
    
def add_sample(samples, x, y, name, show=True):
    samples[name] = (x, y)
    if show: plot_sample(samples, name)
        
def create_cluster_sample(x_list, y_list):
    return np.hstack(x_list), np.hstack(y_list)
```

```python
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=200, n_features=2, centers=5, cluster_std=0.5, random_state=1)
x, y = X[:, 0], X[:, 1]
add_sample(samples, x, y, 'spots')
```

```python
n = 100
x_1 = np.arange(n)
x_2 = np.arange(n) + 0.5
y_1 = 2 * x_1 + np.random.normal(scale=5.0, size=n) + 40
y_2 = 0.5 * x_2 + np.random.normal(scale=8.0, size=n)
x, y = create_cluster_sample([x_1, x_2], [y_1, y_2])
add_sample(samples, x, y, 'two lines')
```

```python
X_1, _ = make_blobs(n_samples=100, n_features=2, centers=5, cluster_std=1.2, random_state=5, center_box=(-25.0, 0.0))
X_2, _ = make_blobs(n_samples=100, n_features=2, centers=3, cluster_std=1.2, random_state=10, center_box=(5.0, 23.0))
X_3, _ = make_blobs(n_samples=100, n_features=2, centers=5, cluster_std=1.2, random_state=20, center_box=(25.0, 45.0))
x, y = create_cluster_sample([X_1[:, 0], X_2[:, 0], X_3[:, 0]], [X_1[:, 1], X_2[:, 1], X_3[:, 1]])
add_sample(samples, x, y, 'hierarchical')
```

```python
X_1, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=3.0, random_state=5, center_box=(-7.0, 0.0))
X_2, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=1.0, random_state=1, center_box=(0.0, 5.0))
x, y = create_cluster_sample([X_1[:, 0], X_2[:, 0]], [X_1[:, 1], X_2[:, 1]])
add_sample(samples, x, y, 'two different spots')
```

```python
x_bridge = np.linspace(-5, 5, 80)
y_bridge = x_bridge + np.random.normal(size=80, scale=.3)
X_1, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=1.0, random_state=5, center_box=(0.0, 0.0))
X_2, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=1.0, random_state=2, center_box=(6.0, 6.0))
X_3, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=1.0, random_state=1, center_box=(-6.0, -6.0))
x, y = create_cluster_sample([X_1[:, 0], X_2[:, 0], X_3[:, 0], x_bridge],
                             [X_1[:, 1], X_2[:, 1], X_3[:, 1], y_bridge])
add_sample(samples, x, y, 'three spots with bridges')
```

```python
x_1 = np.linspace(0, 3, 100)
y_1 = -x_1 + np.random.normal(size=100, scale=.2)
y_2 = x_1 - 4 + np.random.normal(size=100, scale=.2)
x, y = create_cluster_sample([x_1, x_1], [y_1, y_2])
add_sample(samples, x, y, 'cross')
```

```python
x_1 = np.linspace(0, 3, 7)
y_1 = np.linspace(0, 0, 7)
x_2 = np.linspace(0, 3, 7)
y_2 = np.linspace(3, 3, 7)
x_3 = np.linspace(3, 3, 7)
y_3 = np.linspace(0, 3, 7)
x_4 = np.linspace(0, 0, 7)
y_4 = np.linspace(0, 3, 7)
x_0 = np.linspace(-1, 4, 25)
y_0 = (x_0) ** 2 / 6 + 2
x, y = create_cluster_sample([x_0, x_1, x_2, x_3, x_4],
                             [y_0, y_1, y_2, y_3, y_4])
add_sample(samples, x, y, 'square and curve')
```

```python
X_1, _ = make_blobs(n_samples=200, n_features=2, centers=1, cluster_std=1.0, random_state=5, center_box=(-5.0, -5.0))
X_2, _ = make_blobs(n_samples=100, n_features=2, centers=1, cluster_std=6.0, random_state=2, center_box=(-5.0, -5.0))
X_3, _ = make_blobs(n_samples=50, n_features=2, centers=1, cluster_std=15.0, random_state=2, center_box=(-5.0, -5.0))
x, y = create_cluster_sample([X_1[:, 0], X_2[:, 0], X_3[:, 0]], [X_1[:, 1], X_2[:, 1], X_3[:, 1]])
add_sample(samples, x, y, 'no clusters')
```

```python
from sklearn import datasets
np.random.seed(0)
XY = datasets.make_circles(n_samples=600,
                           factor=.5, noise=.05)[0].T
add_sample(samples, *XY, 'concentric circles')
```

```python
XY = datasets.make_moons(n_samples=600, noise=.05)[0].T
add_sample(samples, *XY, 'moons')
```

```python
XY = np.random.rand(600, 2).T
add_sample(samples, *XY, 'uniform')
```

```python
n_dots = 300
t = np.linspace(0, 4 * np.pi, n_dots)
R = np.linspace(1, 6, n_dots)
x_1 = R * np.cos(t)
y_1 = R * np.sin(t)
x_2 = R * np.cos(t + np.pi)
y_2 = R * np.sin(t + np.pi)
x, y = create_cluster_sample([x_1, x_2], [y_1, y_2])
add_sample(samples, x, y, 'spirals')
```

```python
from sklearn.preprocessing import scale
def normalize_sample(x, y):
    return scale(x, with_mean=True, with_std=True), scale(y, with_mean=True, with_std=True)
for sample in samples:
    samples[sample] = normalize_sample(*samples[sample])
```

```python
import pickle
with open('samples.pkl', 'wb') as f:
    pickle.dump(samples, f, pickle.HIGHEST_PROTOCOL)
print(len(samples))
```

```python
for precision in [0., 0.25, 0.5, 0.75, 1.][::-1]:
    recall = np.linspace(0, 1, 101)
    f1_score = 2 * precision * recall / (precision + recall)
    plt.plot(recall, f1_score, label='precision = {}'.format(precision), lw=2)
plt.xlabel('recall')
plt.ylabel('f1-score')
plt.title('f1-score depends on recall')
plt.ylim(-0.1, 1.1)    
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.);
```
