Две основные стратегии:
* [one-vs-one](http://scikit-learn.org/stable/modules/generated/sklearn.multiclass.OneVsOneClassifier.html)
* [one-vs-rest](http://scikit-learn.org/stable/modules/generated/sklearn.multiclass.OneVsRestClassifier.html) (and multilabel task)

Стратегии перехода от бинарного к мультиклассового классификатора. 

one-vs-one - учим 3 классификатора (если 3 класса), каждый класс отличем от каждого. Устраиваем голосование по ответам классификатора. Недостаток - число классификаторов квадратично зависит от числа классов. Если 10 классов, то надо обучить 45 классификаторов (10х9/2).

one-vs-rest - Учим несколько классификаторов по другой схеме. Учим классификатор, которые отличает один класс от остальных (остальны объединяем все в один класс). Количество классификаторов растет линейно. 

Подзадачи: 
- мультилайбелтаск - каждый объект может иметь несколько классов (музыкальный жанр и песни, песни могут быть в несколько жанров). one-vs-rest можем использовать. Все песни жанра пропивопоставляем все, которые не относятся. Ответом будет матрица, строк сколько объектов в тестовой выборке, а столбцы - каждый класс. 

OneVsOneClassifier - на входе алгоритм бинарной классификации, а на выходе алгоритм для мультиклассовой классификации. 
Можно распараллеливать процесс. Все работает независимо. 

OneVsRestClassifier - на вход принимает эстиматор бинарной классификации, а на выходе эстиматор для мультиклассовой классификации. Можно использовать для задач, когда у объекта не один класс. 

Качество в задаче мультиклассовой классификации. 
Точность, насколько правильно все объекты, определенный класс. 
Конфьюжен метрикс - можно визуализировать. По оси х - классы, на самом деле у объектов, по оси у - классы, которые дал наш классификатор.
